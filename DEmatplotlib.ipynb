 Importing all the necessary files:

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
data = pd.read_csv("max_temp.csv")
print(data)
```

```python
df=pd.DataFrame(data)
print(df)
```

```python
days = list(df.iloc[:,0])
city_1 = list(df.iloc[:,1])
city_2 = list(df.iloc[:,2])
city_3 = list(df.iloc[:,3])
city_4 = list(df.iloc[:,4])
```

#FINAL OUTPUT

```python
plt.figure(figsize=(8,6))

plt.plot(days, city_4, marker='o')
plt.plot(days, city_2, marker='o')

font1 = {'family':'serif','color':'olive','size':20}

plt.title("Variation of temperature in cities over a month", fontdict=font1)
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.legend(["city4", "city2"], loc ="upper right")

plt.grid()

plt.show()
```
